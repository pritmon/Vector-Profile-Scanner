# Vector Profile Scanner - Data Exploration

This notebook demonstrates how Data Scientists at Google often perform Exploratory Data Analysis (EDA) on raw datasets before training custom models.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset natively from the internal project structure
dataset_path = 'data/raw/skills.csv'
df = pd.read_csv(dataset_path)

df.head()
```

```python
# Visualize the class distribution
class_counts = df['label'].value_counts()

plt.figure(figsize=(6,4))
class_counts.plot(kind='bar', color=['#4285F4', '#EA4335'])
plt.title('Distribution of Google Skills vs Non-Relevant Skills')
plt.xticks(ticks=[0,1], labels=['Target (1)', 'Non-Target (0)'], rotation=0)
plt.ylabel('Count')
plt.show()
```
